```python
def query_image_annot_ids_contain_obj(obj="zebra"):
    image_annot_id_sets = []
    general_attr_count = {}
    general_attrs = []
    specific_attrs = []
    #object_count = {}
    for i, ele in enumerate(objects):
        if obj in ele:
            image_annot_id_sets.append(annot_ids_to_objects[i])
            # process attributes in the ele
            start_indx = annot_ids_to_objects[i]
            end_indx = annot_ids_to_objects[i+1] if i+1 < len(annot_ids_to_objects) else len(ovad2000['annotations'])
            for ind in range(start_indx, end_indx):
                if ovad2000['categories'][ovad_id2id[ovad2000['annotations'][ind]['category_id']]]['name'] == obj:
                    att_vec = np.array(ovad2000['annotations'][ind]['att_vec']) * retained_attributes
                    for j, attr in enumerate(att_vec):
                        if attr == 1:
                            
                            if ovad2000['attributes'][j]['name'] not in general_attr_count:
                                general_attr_count[ovad2000['attributes'][j]['name']] = 1
                                # object_count[ovad2000['attributes'][j]['name']] = 1
                            else: #if ovad2000['attributes'][j]['name'] in general_attr and ovad2000['attributes'][j]['name'] not in specific_attr:
                                general_attr_count[ovad2000['attributes'][j]['name']] += 1
                                # object_count[ovad2000['attributes'][j]['name']] += 1
                                # specific_attr.append(ovad2000['attributes'][j]['name'])
                                # general_attr.pop(general_attr.index(ovad2000['attributes'][j]['name']))
    
    type_count = {}
    for key_ in list(general_attr_count.keys()):
        key = key_.split(":")[0]
        if 'color' in key:
            # print(key)
            if 'color quantity' in key:
                if 'color quantity' in type_count:
                    # print(key_, "hi")
                    type_count['color quantity'] += 1 
                else:
                    type_count['color quantity'] = 1 
                
        else:            
            if key in type_count:
                type_count[key] += 1 
            else:
                type_count[key] = 1
    
    
    num_color_quantity = general_attr_count['color quantity'] if 'color quantity' in general_attr_count else -1
    for key in list(general_attr_count.keys()):
        # process color
        if 'color:' in key:
            if general_attr_count[key] != num_color_quantity:
                type_count['color'] = 2
                break
                      
    # print(type_count.keys())        
    for key in list(type_count.keys()):
        # print(type_count[key], key)
        if type_count[key] == 1:
            general_attrs.append(key)
        else:
            # print(key)
            specific_attrs.append(key)
        
    return image_annot_id_sets, general_attrs, specific_attrs, general_attr_count#, object_count
```
